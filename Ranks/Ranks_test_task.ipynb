### Анализ и обработка пропущенных значений

```python
data.isna().sum()
```

<div style="background-color: #112335; color:#ffffff; padding: 10px; border-radius: 10px;">
    
В данных много пропущенных значений. Посмотрим, есть ли какая-то связь, объясняющая данное явление.

```python
# сгруппируем компании по капитализации
data.capitalization.unique()
```

```python
# посчитаем общее количество ячеек и пропущенных значений
total_values = data.count().sum()
na_total = data.isna().sum().sum()

for group in ['Крупнейшие',
              'Крупные', 
              'Средние', 
              'Малые',
              'Микро',
              'Нано']:
    
    # посчитаем "вес" группы и сколько в ней пропущено значений
    na_group = data[data.capitalization == group].isna().sum().sum()
    na_group_total_perc = na_group / na_total * 100
    
    group_values = data[data.capitalization == group].count().sum()
    group_total_perc = group_values / total_values * 100
    
    na_group_perc = na_group / group_values * 100
    print('------------')
    print(f'Группа "{group}": доля данных относительно всех данных - {round(group_total_perc, 2)}%.',
          f'Доля пропусков среди всех пропусков - {round(na_group_total_perc, 2)}%.',
          f'Доля пропусков внутри группы - {round(na_group_perc, 2)}%')
```

<div style="background-color: #112335; color:#ffffff; padding: 10px; border-radius: 10px;">
    
Опреленная зависимость присутствует:

1. `Количество пропусков для каждой подгруппы по капитализации коррелирует с общим количеством данных для группы.` Исключение - группа "Малые" (доля пропусков больше, чем общая доля значений).
2. Также можно заметить, что `чем меньше компания по капитализации, тем меньше информации собрано о ней`. Минимум - у групп "Микро" и "Нано" (порядка 45%).
    
Можно сделать вывод: данные пропущены не из-за утечки, а из-за особенностей сбора для каждой компании - либо для части компаний данные не собираются, либо не предоставляются компаниями вовсе.

### Обсуждение подхода исследования

<div style="background-color: #112335; color:#ffffff; padding: 10px; border-radius: 10px;">
    
Для анализа данных создадим сводную таблицу для каждого категориального признака. В такой таблице будут отражены изменение цены и ранков. Далее выберем наилучшую корреляцию для признаков и оценим её с помощью коэффициента Пирсона. Силу корреляции классифицируем с помощью шкалы Чеддока. Далее построим графики и убедимся, что корреляция по выбранным признакам присутствует.

<div style="background-color: #027AE8; color:#ffffff; padding: 15px; border-radius: 10px">
    
### Процесс исследования

```python
# составим 2 списка признаков для последующего построения графиков
rank_list = ['credit_rank',
             'forecast_rank',
             'growth_rank', 
             'value_rank',
             'esg_rank',
             'price_momentum_rank',
             'industry_rank',
             'rank_4',
             'rank_5',
             'rank_6',
             'rank_8']

features_list = ['perf_total',
                 'perf_q',           
                 'perf_1y',
                 'perf_3y',
                 'perf_5y',
                 'credit_rank',
                 'forecast_rank',
                 'growth_rank', 
                 'value_rank',
                 'esg_rank',
                 'price_momentum_rank',
                 'industry_rank',
                 'rank_4',
                 'rank_5',
                 'rank_6',
                 'rank_8']
```

<div style="background-color: #112335; color:#ffffff; padding: 10px; border-radius: 10px;">
    
Возьмем срез по континентам.

```python
# сгруппируем данные по дате и признаку "континент"
continent_group = data.groupby(['date_round', 'continent'])[features_list].mean().reset_index()
continent_group
```

```python
continent_group.continent.unique()
```

<div style="background-color: #112335; color:#ffffff; padding: 10px; border-radius: 10px;">
    
Возьмем данные одного континента и составим сетку графиков, где попарно будут отображены изменение цены по периодам и изменение ранков по блокам.

```python
# сгруппируем данные по значению "континента"
africa = continent_group[continent_group.continent == 'Africa'].reset_index(drop=True)
for rank in rank_list:
    
    fig, axs = plt.subplots(1, 5, figsize=(20, 3))
    
    axs[0].plot(africa.date_round, africa.perf_total)
    axs[0].plot(africa.date_round, africa[rank])
    axs[0].set_title(f'perf_total / {rank}')
    
    axs[1].plot(africa.date_round, africa.perf_q)
    axs[1].plot(africa.date_round, africa[rank])
    axs[1].set_title(f'perf_q / {rank}')
    
    axs[2].plot(africa.date_round, africa.perf_1y)
    axs[2].plot(africa.date_round, africa[rank])
    axs[2].set_title(f'perf_1y / {rank}')
    
    axs[3].plot(africa.date_round, africa.perf_3y)
    axs[3].plot(africa.date_round, africa[rank])
    axs[3].set_title(f'perf_3y / {rank}')
    
    axs[4].plot(africa.date_round, africa.perf_5y)
    axs[4].plot(africa.date_round, africa[rank])
    axs[4].set_title(f'perf_5y / {rank}')
    
    # выведем сетку простых графиков, а далее детально рассмотрим графики с лучшей корреляцией
    plt.show()
```

<div style="background-color: #112335; color:#ffffff; padding: 10px; border-radius: 10px;">
    
Можно заметить некую корреляцию на графике `perf_5y` и `forecast_rank` - запомним данные параметры. Изучим графики по всем континентам. Получим некоторые зависимости:

1. Африка: `perf_5y` и `forecast_rank`.
2. Европа: `perf_total` и `value_rank`.
3. Океания: `perf_total` и `esg_rank`.

Отобразим их отдельно.

```python
europe = continent_group[continent_group.continent == 'Europe'].reset_index(drop=True)
oceania = continent_group[continent_group.continent == 'Oceania'].reset_index(drop=True)

fig, axs = plt.subplots(1, 3, figsize=(12, 3))

axs[0].plot(africa.date_round, africa.perf_5y, label='perf_5y')
axs[0].plot(africa.date_round, africa.forecast_rank, label='forecast_rank')
axs[0].set_title(f'Africa: perf_5y / forecast_rank')
axs[0].set_ylabel('Изменение цены / Ранк')
axs[0].set_xlabel('Дата отсчета')
axs[0].legend()

axs[1].plot(europe.date_round, europe.perf_total, label='perf_total')
axs[1].plot(europe.date_round, europe.value_rank, label='value_rank')
axs[1].set_title(f'Europe: perf_total / value_rank')
axs[1].set_ylabel('Изменение цены / Ранк')
axs[1].set_xlabel('Дата отсчета')
axs[1].legend()

axs[2].plot(oceania.date_round, oceania.perf_total, label='perf_total')
axs[2].plot(oceania.date_round, oceania.esg_rank, label='esg_rank')
axs[2].set_title(f'Oceania: perf_total / esg_rank')
axs[2].set_ylabel('Изменение цены / Ранк')
axs[2].set_xlabel('Дата отсчета')
axs[2].legend()

plt.show()
```

<div style="background-color: #112335; color:#ffffff; padding: 10px; border-radius: 10px;">
    
Выдвинем гипотезу: корреляция для указанных параметров существует. Найдем силу корреляции с помощью коэффициента Пирсона.

```python
africa_corr = africa.perf_5y.corr(africa.forecast_rank)
display(africa_corr)

europe_corr = europe.perf_total.corr(europe.value_rank)
display(europe_corr)

oceania_corr = oceania.perf_total.corr(oceania.esg_rank)
display(oceania_corr)
```
